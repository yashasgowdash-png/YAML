```python
# Cell 1: Import libraries
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Set plot style
sns.set_style('whitegrid')
```

```python
# Cell 2: Load Titanic dataset
titanic = sns.load_dataset('titanic')

# Show first few rows
titanic.head()
```

```python
# Cell 3: Basic info and summary stats
print("Dataset info:")
print(titanic.info())

print("\nSummary statistics:")
print(titanic.describe(include='all'))
```

```python
# Cell 4: Check missing values
missing = titanic.isnull().sum()
print("Missing values per column:")
print(missing)
```

```python
# Cell 5: Visualize count of survivors
plt.figure(figsize=(6,4))
sns.countplot(data=titanic, x='survived')
plt.title('Survival Count')
plt.xlabel('Survived (0=No, 1=Yes)')
plt.ylabel('Count')
plt.show()
```

```python
# Cell 6: Survival rate by sex
plt.figure(figsize=(6,4))
sns.barplot(data=titanic, x='sex', y='survived')
plt.title('Survival Rate by Sex')
plt.ylabel('Survival Rate')
plt.show()
```

```python
# Cell 7: Survival rate by passenger class
plt.figure(figsize=(6,4))
sns.barplot(data=titanic, x='class', y='survived')
plt.title('Survival Rate by Passenger Class')
plt.ylabel('Survival Rate')
plt.show()
```

```python
# Cell 8: Age distribution by survival
plt.figure(figsize=(8,5))
sns.histplot(data=titanic, x='age', hue='survived', multiple='stack', bins=30)
plt.title('Age Distribution by Survival')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
# Select only numeric columns for correlation
numeric_cols = titanic.select_dtypes(include=['number'])

plt.figure(figsize=(8,6))
sns.heatmap(numeric_cols.corr(), annot=True, cmap='coolwarm')
plt.title('Feature Correlation Heatmap')
plt.show()
```
